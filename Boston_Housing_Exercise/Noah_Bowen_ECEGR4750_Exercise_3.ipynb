```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

dataset = pd.read_csv('Boston.csv')
dataset.head()

nox_np = np.array(dataset.nox)
indus_np = np.array(dataset.indus)
print("Correlation Coefficient = ", np.corrcoef(nox_np, indus_np))
y = nox_np
X = indus_np
X = sm.add_constant(X)

lr_model = sm.OLS(y,X).fit()
print(lr_model.summary())
X_prime = np.linspace(X.min(), X.max(), 100)
X_prime = sm.add_constant(X_prime)
y_hat = lr_model.predict(X_prime)

plt.scatter(indus_np,nox_np)
# Add the regression line (in red).
plt.plot(X_prime[:,1], y_hat, "red", alpha = 0.9)
plt.xlabel("Industry")
plt.ylabel("Noxious Gases")
```

<p>I chose to set the independent variable when looking at noxious gas levels to be based on the % of land taken up by industry because they are more likely to produce noxious gas than homes</p>

<h2>Part 2: MEDV Linear Regression</h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

dataset = pd.read_csv('Boston.csv')
dataset.head()

medv_np = np.array(dataset.medv)
lstat_np = np.array(dataset.lstat)
print("Correlation Coefficient = ", np.corrcoef(medv_np,lstat_np))
y = medv_np
X = lstat_np
X = sm.add_constant(X)

lr_model = sm.OLS(y,X).fit()
print(lr_model.summary())
X_prime = np.linspace(X.min(), X.max(), 100)
X_prime = sm.add_constant(X_prime)
y_hat = lr_model.predict(X_prime)

plt.scatter(lstat_np,medv_np)
# Add the regression line (in red).
plt.plot(X_prime[:,1], y_hat, "red", alpha = 0.9)
plt.xlabel("Median Value")
plt.ylabel("% Lower Status")
```

<p>I chose to set the independent variable when looking at median value of homes to be % of residents of lower status because those of lower status are less likely to own expensive homes and it had the best correlation coefficient</p>
